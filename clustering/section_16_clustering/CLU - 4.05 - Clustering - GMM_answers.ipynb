## Clustering - Gaussian Mixed Models

In this notebook we'll:
1. Look at failure case for K-means
2. Use the same example for GMM

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
X, _ = make_blobs(n_samples=1500,
                  cluster_std=[1.5, 1.5, 1.5],
                  random_state=4)

X = np.dot(X, [[0.6, -0.6], [-0.4, 0.8]])

plt.scatter(X[:, 0], X[:, 1], s=10, marker='o', facecolors='none', edgecolors='black')
plt.title("Ellipsis Cluster")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
kmeans = KMeans(n_clusters=2).fit(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
plt.title("Kmeans Failure case")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=2, covariance_type='full').fit(X)
labels = gmm.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title("GMM")
plt.xticks([])
plt.yticks([])
plt.show()
```

### Sampling from a GMM

```python
sample_X, sample_labels = gmm.sample(300)
plt.scatter(sample_X[:, 0], sample_X[:, 1], c=sample_labels)
plt.title("GMM Sample 1")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
sample_X, sample_labels = gmm.sample(300)
plt.scatter(sample_X[:, 0], sample_X[:, 1], c=sample_labels)
plt.title("GMM Sample  2")
plt.xticks([])
plt.yticks([])
plt.show()
```
